```python
import torch
import random
import numpy as np
from collections import deque
```

```python
memory = deque(maxlen=10000)

for i in range(10):
    state = torch.rand([3,300,300])
    next_state = torch.rand([3,300,300])
    action = torch.randint(10,[1])
    reward = torch.randint(10,[1])
    done = torch.tensor([True])
    memory.append((state, next_state, action, reward, done))
```

```python
batch = random.sample(memory, 5)
```

```python
for i in zip(*batch):
    print(i[0])
    break
```
